```python

#enter information to be posted in AI

paragraph = input("Enter the paragraph to be masked:\n")


#enter List of words to be hide

user_input = input("\nEnter words to mask (comma-separated): ")
words_to_mask = [w.strip() for w in user_input.split(",")]

#Reversible mask function
import re

def reversible_mask(text, words):

    mapping = {}
    masked_text = text

    for i, word in enumerate(words, start=1):
        if word:
            mask = f"[MASK_{i}]"
            masked_text = re.sub(
                re.escape(word),
                mask,
                masked_text,
                flags=re.IGNORECASE
            )
            mapping[mask] = word

    return masked_text, mapping
#masking
masked_text, mask_mapping = reversible_mask(paragraph, words_to_mask)

print("\n--- Masked Text ---")
print(masked_text)

print("\n--- Mask Mapping (store securely) ---")
print(mask_mapping)

#assume AI Responce
response_text = input(
    "\nEnter a response text that uses the same masks:\n"
)

#unmask function
def unmask(text, mapping):
    unmasked_text = text
    for mask, original in mapping.items():
        unmasked_text = unmasked_text.replace(mask, original)
    return unmasked_text
#unmask
final_output = unmask(response_text, mask_mapping)

print("\n--- Unmasked Response ---")
print(final_output)



```
